```python
#Leitura da base de dados em csv.
#Utilizado o sep para organizar e separar os dados
#E o head retorna as 5 primeiras linhas

df = pd.read_csv('DIMIF_desmatamento.csv', sep=';')
df.head()
```

# 1º Pergunta - Qual UC tem a maior área estimada (ha) ?, use gráfico e tabela para demonstrar;

```python
#Irei utilizar o iloc para selecionar linhas e colunas por número, e assim transformar todos os valores NAN do dataset em 0

df.iloc[:,0:33]=df.iloc[:,0:33].fillna(0)
df
```

```python
#Na base de dados existe uma linha 0 e a mesma está com vários tipos de dados errado e com isso implicará na hora de se obter 
#resultados, utilizarei a função drop para remover essa linha

df.drop(0, inplace=True)
```

```python
#Converter para numérico o tipo de dados forçando no modo "coerce".

df['Área estimada da UC (ha)']=pd.to_numeric(df['Área estimada da UC (ha)'], errors="coerce")
```

```python
#Visto que a área estimada foi convertida de object para float

df.dtypes
```

```python
#Criei um df2 e aloquei a base de dados df, dei iloc para localizar as linhas e colunas, e no final coloquei que só pode aparecer
#os valores que for acima de 700.000

df2 = df.loc[df['Área estimada da UC (ha)'] > 700.000]
```

```python
#Retorna os valores exclusivos de Nome da UC

df2['Nome da UC'].unique()
```

```python
#Irei agrupar na base de dados a área estimada com o nome da UC e assim visualizar qual UC tem a maior área estimada(ha)

df3 = df2['Área estimada da UC (ha)'].groupby(df2['Nome da UC']).sum()
df3
```

```python
#Criar uma dataframe com os dados da área (ha) e do nome da UC e ordenar as informações em ordem crescente.

df4 = pd.DataFrame(df3.sort_values())
df4
```

```python
#Assim gerando um gráfico para se entender que a UC de FLONA de Pau rosa tem a maior área estimada (ha)

df4.plot.bar()
```

# Resposta - A UC com maior área estimada é a de FLONA de Pau-Rosa

# 4º Pergunta - Construa uma tabela que relacione estados e área desmatada, respondendo qual os estados que mais e o que menos teve área desmatada nos anos computados na pesquisa.

```python
#Irei criar uma coluna no dataframe com o nome Total Desmatado
#Depois localizar as linhas e colunas desde o 10 até o final 33 e assim gerando o axis=1 (eixo) que produz uma soma para cada linha

df['Total Desmatamento'] = df.iloc[:, 10:33].sum(axis=1)
df.head()
```

```python
#Retorna os valores exclusivos

df['Coordenação Regional do ICMBio'].unique()
```

```python
#Irei agrupar na base de dados a coluna criada Total Desmatada com a coordenação e somar os valores
df4 = df['Total Desmatamento'].groupby(df['Coordenação Regional do ICMBio']).sum()
df4
```

```python
#Criar um dataframe com a soma das informações da base anterior e colocar os valores em ordem crescente

df5 = pd.DataFrame(df4.sort_values())
df5
```

```python
##Gerar um gráfico do tipo barra em horizontal

df5.plot.barh()
```

# Resposta - Os estados com menos área desmatada foi Cabedelo-PB e Florianopólis-SC e com maior área devastada foi Santarém-PA e Porto Velho-RO.
