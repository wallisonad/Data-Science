# 3° Pergunta - Monte um gráfico da quantidade de área queimada por ano, mostrando a evolução das queimadas

```python
#Primeiramente iremos checar os tipos de dados das áreas de queimadas, visto que está como Objeto, será feito a conversão de objeto para float

dados[['Área queimada em 2012','Área queimada em 2013','Área queimada em 2014','Área queimada em 2015','Área queimada em 2016','Área queimada em 2017','Área queimada em 2018']]
dados.head()
```

```python
# A soma das quantidades de valores nulos (NAN)

dados.isnull().sum()
```

```python
#Converter o que for (vírgula) por (ponto)

dados['Área queimada em 2012'] = dados['Área queimada em 2012'].str.replace(',','.')
dados['Área queimada em 2013'] = dados['Área queimada em 2013'].str.replace(',','.')
dados['Área queimada em 2014'] = dados['Área queimada em 2014'].str.replace(',','.')
dados['Área queimada em 2015'] = dados['Área queimada em 2015'].str.replace(',','.')
dados['Área queimada em 2016'] = dados['Área queimada em 2016'].str.replace(',','.')
dados['Área queimada em 2017'] = dados['Área queimada em 2017'].str.replace(',','.')
dados['Área queimada em 2018'] = dados['Área queimada em 2018'].str.replace(',','.')
```

```python
#Converter para numérico o tipo de dados forçando no modo "coerce".

dados['Área queimada em 2018']=pd.to_numeric(dados['Área queimada em 2018'], errors="coerce")
dados['Área queimada em 2017']=pd.to_numeric(dados['Área queimada em 2017'], errors="coerce")
dados['Área queimada em 2016']=pd.to_numeric(dados['Área queimada em 2016'], errors="coerce")
dados['Área queimada em 2015']=pd.to_numeric(dados['Área queimada em 2015'], errors="coerce")
dados['Área queimada em 2014']=pd.to_numeric(dados['Área queimada em 2014'], errors="coerce")
dados['Área queimada em 2013']=pd.to_numeric(dados['Área queimada em 2013'], errors="coerce")
dados['Área queimada em 2012']=pd.to_numeric(dados['Área queimada em 2012'], errors="coerce")
dados['Área estimada da UC (ha)']=pd.to_numeric(dados['Área estimada da UC (ha)'], errors="coerce")
```

```python
#Mostra os tipos de dados

dados.dtypes
```

```python
#Irei utilizar o iloc para selecionar linhas e colunas por número, e assim transformar todos os valores NAN do dataset em 0

dados.iloc[:, 9:16]=dados.iloc[:, 9:16].fillna(0)
dados.head()
```

```python
#Irei utilizar o iloc para selecionar linhas e colunas por número e somar os valores

dados2= dados.iloc[:, 9:16].sum()
```

```python
#Agora irei criar um dataframe utilizando os dados anteriores e criando mais uma coluna com o nome Queimadas, colocando os valores em ordem crescente através do sort.values

dados3= pd.DataFrame(dados2.sort_values())
dados3.columns=['Queimadas']
dados3
```

```python
#Gerado gráfico para identificar a evolução das queimadas  através do plot em barras na horizontal

dados3.plot.barh()
```

# 4º Pergunta -  Construa uma tabela que relacione estados e área queimada, respondendo qual o estado que mais e o que menos teve área queimada nos anos computados na pesquisa.

```python
#Irei criar uma coluna no dataframe com o nome Total Desmatado
#Depois localizar as linhas e colunas desde o 9 até o final 16 e assim gerando o axis=1 (eixo) que produz uma soma para cada linha

dados['Total Queimada'] = dados.iloc[:, 9:16].sum(axis=1)
dados.head()
```

```python
#Retorna os valores exclusivos

dados['Coordenação Regional do ICMBio'].unique()
```

```python
#Agora vamos agrupar na base de dados a coluna criada Total Desmatada com a coordenação e somar os valores

dados4 = dados['Total Queimada'].groupby(dados['Coordenação Regional do ICMBio']).sum()
dados4
```

```python
#Irei criar um dataframe utilizando os dados anteriores e colocando os valores gerados em ordem crescente

dados5 = pd.DataFrame(dados4.sort_values())
dados5
```

```python
#Gerado o gráfico mostrando qual o estado que mais teve área queimada e o que menos teve área queimada.

dados5.plot.barh()
```
